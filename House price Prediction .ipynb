```python
x_null_index=x.isnull().sum()[x.isnull().sum()>0].index
```

```python
x[x_null_index[0]].dtype=='float64'
```

```python
def null_handler(ind,df):
    int_list=[] 
    obj_list=[] 
    extra=[]
    x=df
    for i in ind: 
        if x[i].dtype=='float64' or x[i].dtype=='int64': 
            x[i]=x[i].fillna(x[i].mean()) 
            int_list.append(i)
        elif x[i].dtype=='object':
            x[i]=x[i].fillna(x[i].mode()[0]) 
            obj_list.append(i)
        else: extra.append(i) 
    print(int_list,len(int_list)) 
    print(obj_list,len(obj_list)) 
    print(extra,len(extra)) #print(x.isnull().sum())
    return x
```

```python
df_x=null_handler(x_null_index,x)

df_x.isnull().sum()[df_x.isnull().sum()>0]
```
